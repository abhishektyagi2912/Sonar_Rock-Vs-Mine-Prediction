```python
X = df.drop(columns=60, axis=1) ## axis=1 means column wise and axis=0 means row wise
X.head()
```

```python
Y = df[60]
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1) ## stratify is used to maintain the ratio of data
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_train, model.predict(X_train)) ## 0.8347107438016529 means 83.47% accuracy
```

```python
accuracy_score(Y_test, model.predict(X_test)) ## 0.7619047619047619 means 76.19% accuracy
```
